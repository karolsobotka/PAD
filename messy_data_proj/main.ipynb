```python
import pandas as pd
data = pd.read_csv('messy_data.csv', na_values='NaN')
#data_no_duplicates = data.drop_duplicates()

for index, row in data.iterrows():
    for column_name, value in row.items():
        if data.at[index, column_name] == ' ':
                  data.at[index, column_name] = 'NAN'
        else:
            data.at[index, column_name] = str(value).upper()
data.head(5)

```

```python
columns = ['carat', ' clarity',  ' color', ' cut', ' x dimension', ' y dimension', ' z dimension',
                          ' depth', ' table', ' price']
for column in columns:
    print(f'{str(column).strip()}:')
    print(f'ile braków danych: {sum(data[column].isna())}')
    if column not in [ 'carat', 'color', 'cut','table']:
        print(f'unikalne wartości: {data[column].value_counts()}')
    else:
        print(f'unikalne wartości: {data[column].nunique()}')
    #separator lini bo brzydko wygląda
    print()
```
