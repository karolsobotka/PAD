```python
import pandas as pd
data = pd.read_csv('messy_data.csv', na_values='NAN', )
data.head(200)
```

```python
columns = ['carat', ' clarity',  ' color', ' cut', ' x dimension', ' y dimension', ' z dimension',
                          ' depth', ' table', ' price']
for column in columns:
    print(f'{column} :')
    print(f'ile braków danych: {sum(data[column].isna())}')
    if column not in [ 'carat', 'color', 'cut','table']:
        print(f'unikalne wartości: {data[column].value_counts()}')
    else:
        print(f'unikalne wartości: {data[column].nunique()}')
```
